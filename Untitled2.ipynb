# Questão 1 — Classificação de Temperaturas (Condicionais + Loop)

Dada a lista de temperaturas registradas durante a semana:
temperaturas = [18, 22, 30, 35, 28, 15, 40]
Classifique cada dia como:
•	Abaixo de 20 → "Frio"
•	Entre 20 e 30 → "Agradável"
•	Acima de 30 → "Quente"


1.	Criar uma lista com as classificações

```python
temperaturas = [18, 22, 30, 35, 28, 15, 40]
classificacoes = []

for temp in temperaturas:
    if temp < 20:
        classificacoes.append(f'{temp}°C: Frio')
    elif 20 <= temp <= 30:
        classificacoes.append(f'{temp}°C: Agradável')
    else:
        classificacoes.append(f'{temp}°C: Quente')

for classificacao in classificacoes:
    print(classificacao)
```

2.	Contar quantos dias houve em cada categoria

```python
contagem_categorias = {'Frio': 0, 'Agradável': 0, 'Quente': 0}

for classificacao_str in classificacoes:
    if 'Frio' in classificacao_str:
        contagem_categorias['Frio'] += 1
    elif 'Agradável' in classificacao_str:
        contagem_categorias['Agradável'] += 1
    elif 'Quente' in classificacao_str:
        contagem_categorias['Quente'] += 1

print("Contagem de dias por categoria:")
for categoria, contagem in contagem_categorias.items():
    print(f"{categoria}: {contagem} dias")
```

# Questão 2 — Sistema de Avaliação de Alunos

Dada a lista de notas finais:
notas = [4.5, 6.0, 7.8, 9.0, 5.5, 8.2]
Classifique:
•	< 5 → "Reprovado"
•	5 a 7 → "Recuperação"
•	7 → "Aprovado"

1.	Gerar lista de classificações

```python
notas = [4.5, 6.0, 7.8, 9.0, 5.5, 8.2]
classificacoes_notas = []

for nota in notas:
    if nota < 5:
        classificacoes_notas.append(f'{nota}: Reprovado')
    elif 5 <= nota < 7:
        classificacoes_notas.append(f'{nota}: Recuperação')
    else:
        classificacoes_notas.append(f'{nota}: Aprovado')

for classificacao in classificacoes_notas:
    print(classificacao)
```

2.	Calcular:
	Quantidade de aprovados
	/Percentual de reprovados


```python
quantidade_aprovados = 0
quantidade_reprovados = 0

for classificacao in classificacoes_notas:
    if 'Aprovado' in classificacao:
        quantidade_aprovados += 1
    elif 'Reprovado' in classificacao:
        quantidade_reprovados += 1

total_alunos = len(classificacoes_notas)
percentual_reprovados = (quantidade_reprovados / total_alunos) * 100 if total_alunos > 0 else 0

print(f"Quantidade de Aprovados: {quantidade_aprovados}")
print(f"Percentual de Reprovados: {percentual_reprovados:.2f}%")
```

# Questão 3 — Monitoramento de Consumo de Energia

Consumo diário (kWh):
consumo = [120, 95, 200, 180, 75, 220, 160]
Regras:
•	< 100 → "Baixo"
•	100–180 → "Moderado"
•	180 → "Alto"


1.	Filtrar dias com consumo alto




```python
consumo = [120, 95, 200, 180, 75, 220, 160]

def classificar(v):
    if v < 100:   return 'Baixo'
    if v <= 180:  return 'Moderado'
    return 'Alto'

# Filtrar dias com consumo alto
dias_altos = [(i+1, v) for i, v in enumerate(consumo) if classificar(v) == 'Alto']

print("Dias com consumo alto:")
for dia, v in dias_altos:
    print(f"  Dia {dia}: {v} kWh")
```

2.	Calcular:
o	Total de consumo da semana
/	Média semanal


```python
consumo = [120, 95, 200, 180, 75, 220, 160]

total_consumo = sum(consumo)
media_semanal = total_consumo / len(consumo)

print(f"Total de consumo da semana: {total_consumo} kWh")
print(f"Média semanal de consumo: {media_semanal:.2f} kWh")
```

3.	Exibir alerta se houver mais de 2 dias com consumo alto

```python
consumo = [120, 95, 200, 180, 75, 220, 160]

def classificar(v):
    if v < 100:   return 'Baixo'
    if v <= 180:  return 'Moderado'
    return 'Alto'

dias_altos = [(i+1, v) for i, v in enumerate(consumo) if classificar(v) == 'Alto']

if len(dias_altos) > 2:
    print(f"⚠️ ALERTA: {len(dias_altos)} dias com consumo alto!")
    for dia, v in dias_altos:
        print(f"  Dia {dia}: {v} kWh")
else:
    print(f"✅ Normal: apenas {len(dias_altos)} dia(s) com consumo alto.")
```

# Questão 4 — Simulação de Carrinho de Compras


Lista de preços:
precos = [50, 120, 30, 200, 80, 15]
Regras de desconto:
•	Item < 50 → 5% desconto
•	50–150 → 10% desconto
•	150 → 15% desconto


1.	Criar lista com preços finais

```python
precos = [50, 120, 30, 200, 80, 15]

def calcular_desconto(p):
    if p < 50:    return p * 0.95   # 5% desconto
    if p <= 150:  return p * 0.90   # 10% desconto
    return p * 0.85                 # 15% desconto

precos_finais = [calcular_desconto(p) for p in precos]

print("Preços finais:")
for i, (original, final) in enumerate(zip(precos, precos_finais), 1):
    print(f"  Item {i}: R${original:.2f} -> R${final:.2f}")
```

2.	Calcular:
o	Total da compra com desconto
/Economia total obtida


```python
total_original = sum(precos)
total_com_desconto = sum(precos_finais)
economia_total = total_original - total_com_desconto

print(f"Total da compra (sem desconto): R${total_original:.2f}")
print(f"Total da compra (com desconto): R${total_com_desconto:.2f}")
print(f"Economia total obtida: R${economia_total:.2f}")
```
